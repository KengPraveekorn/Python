```python
#การเพิ่มข้อมูล/การเพิ่มแถว
import pandas as pd
df = pd.read_excel("Stock.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
products =[
    ["หูฟัง","อุปกรณ์คอม",400,20,500],
    ["สายชาร์จ","อุปกรณ์คอม",1400,20,1500]
]
cols=["Name","Group","Cost","Quantity","Summation"]
newdf=pd.DataFrame(data=products,columns=cols)
newdf
```

```python
newdf.set_index(["Name"],inplace=True)
newdf
```

```python
df=df.append(newdf) #เอาข้อมูลเดิมไปต่อกับข้อมูลก้อนใหม่
df
```
