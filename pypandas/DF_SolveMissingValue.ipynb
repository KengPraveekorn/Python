```python
#วิธีจัดการข้อมูลที่สูญหาย
# 1.แทนค่าข้อมูลด้วยค่าเฉลี่ย
import pandas as pd
df = pd.read_excel("Employee.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
df.describe() #show ค่าทางสถิติ
```

```python
select = 'Salary'
df[select]=df[select].fillna(df[select].mean()) #ทำงานเฉพาะคอลัมน์ salary และเติมค่าเฉลี่ยของคอลัมน์ salarลงไปแทนที่ค่าว่าง
df
```

```python
# 2.แทนค่าข้อมูลด้วยค่าที่กำหนด
df = pd.read_excel("Employee.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
select = 'Salary'
df[select] = df[select].fillna(18000) #ข้อมูลที่เป็นค่าว่างในคอลัมน์ salary จะถูกแทนที่ด้วย 18000
df
```

```python
# 3.แทนค่าข้อมูลด้วยค่าก่อนหน้า
df = pd.read_excel("Employee.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
df.fillna(method="pad",inplace=True) #แทนที่ค่าว่างด้วยค่าก่อนหน้าทุกคอลัมน์ ยกเว้นคอลัมน์ที่ set เป็น index
df
```

```python
# 4.แทนค่าข้อมูลด้วยค่าถัดไป
df = pd.read_excel("Employee.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
df.fillna(method="bfill",inplace=True) #แทนที่ค่าว่างด้วยค่าถัดไปทุกคอลัมน์ ยกเว้นคอลัมน์ที่ set เป็น index
df
```

```python
# 5.วิธีจัดการข้อมูลที่สูญหายด้วยกาลบ
df = pd.read_excel("Employee.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
# 5.1 ลบแถวทั้งหมดที่มี่ค่าว่าง
df.dropna(inplace=True)
df
```

```python
# 5.2 ลบแถวบางส่วนที่มี่ค่าว่าง
df = pd.read_excel("Employee.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
df.dropna(subset=["Job","Salary"],inplace=True)
df
```

```python
# 5.3 ลบคอลัมน์ที่ข้อมูลไม่ครบออกไป
df = pd.read_excel("Employee.xlsx")
df.set_index(["Name"],inplace=True)
df
```

```python
df.dropna(axis="columns",inplace=True)
df
```
